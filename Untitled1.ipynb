```python
import pandas as pd

# Load the dataset
df = pd.read_csv("students.csv")

# 1. Print the first 5 rows
print("--- First 5 rows ---")
print(df.head())

# 2. Print the shape and data types
print("\n--- Shape ---")
print(df.shape)

print("\n--- Data Types ---")
print(df.dtypes)

# 3. Print summary statistics for numeric columns
print("\n--- Summary Statistics ---")
print(df.describe())

# 4. Print the count of students who passed and failed
print("\n--- Passed / Failed Counts ---")
print(df['passed'].value_counts())

# 5. Compute average score per subject for passing and failing students
subject_cols = ['math', 'science', 'english', 'history', 'pe']

print("\n--- Average scores for PASSING students ---")
print(df[df['passed'] == 1][subject_cols].mean())

print("\n--- Average scores for FAILING students ---")
print(df[df['passed'] == 0][subject_cols].mean())

# 6. Find the student with the highest overall average across all 5 subjects
df['avg_score'] = df[subject_cols].mean(axis=1)
highest_student = df.loc[df['avg_score'].idxmax()]

print("\n--- Student with Highest Overall Average ---")
print(f"Name: {highest_student['name']}")
print(f"Average Score: {highest_student['avg_score']}")
```

```python
#TASK2:
import pandas as pd
import matplotlib.pyplot as plt

# Loading dataset
df = pd.read_csv("students.csv")

# Setup: Add avg_score column
subject_cols = ['math', 'science', 'english', 'history', 'pe']
df['avg_score'] = df[subject_cols].mean(axis=1)

# ==========================================
# Plot 1: Bar Chart
# ==========================================
plt.figure(figsize=(8, 5))
avg_per_subject = df[subject_cols].mean()
plt.bar(avg_per_subject.index, avg_per_subject.values, color='skyblue', edgecolor='black')

plt.title('Average Score per Subject Across All Students')
plt.xlabel('Subject')
plt.ylabel('Average Score')
plt.savefig('plot1_bar.png', bbox_inches='tight')
plt.show()

# ==========================================
# Plot 2: Histogram
# ==========================================
plt.figure(figsize=(8, 5))
plt.hist(df['math'], bins=5, color='lightgreen', edgecolor='black')
mean_math = df['math'].mean()

plt.axvline(mean_math, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {mean_math:.1f}')
plt.title('Distribution of Math Scores')
plt.xlabel('Math Score')
plt.ylabel('Number of Students')
plt.legend()
plt.savefig('plot2_hist.png', bbox_inches='tight')
plt.show()

# ==========================================
# Plot 3: Scatter Plot
# ==========================================
plt.figure(figsize=(8, 5))
passed_df = df[df['passed'] == 1]
failed_df = df[df['passed'] == 0]

plt.scatter(passed_df['study_hours_per_day'], passed_df['avg_score'], color='blue', label='Pass', alpha=0.7, s=80)
plt.scatter(failed_df['study_hours_per_day'], failed_df['avg_score'], color='orange', label='Fail', alpha=0.7, s=80)

plt.title('Study Hours vs Average Score')
plt.xlabel('Study Hours per Day')
plt.ylabel('Average Overall Score')
plt.legend()
plt.savefig('plot3_scatter.png', bbox_inches='tight')
plt.show()


# ==========================================
# Plot 5: Line Plot
# ==========================================
plt.figure(figsize=(10, 5))
plt.plot(df['name'], df['math'], marker='o', linestyle='-', color='purple', label='Math')
plt.plot(df['name'], df['science'], marker='s', linestyle='--', color='teal', label='Science')

plt.title('Math and Science Scores per Student')
plt.xlabel('Student Name')
plt.ylabel('Score')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout() # Ensures student names fit nicely at the bottom
plt.savefig('plot5_line.png')
plt.show()
```
